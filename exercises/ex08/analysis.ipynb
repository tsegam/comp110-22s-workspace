Step 2: 
- The function list_to_int converts the list into the list values to integer. 
- Then I take tutoring effective and oh effective as arrays 
- The Total_student field calculates the total number of students. 
- I then take a count of effetiveness rated by 5 and above and devide that with the total number of students to get a proportion.

```python
def list_to_int(list):
    new_list = []

    for i in list:
        if i == '':
            continue
            
        new_list.append(int(i))
    return new_list

tutoring_effective = list_to_int(selected_data['tutoring_effective'])
oh_effective = list_to_int(selected_data['oh_effective'])

total_students = max(len(tutoring_effective), len(oh_effective))

tutoring_count = count(tutoring_effective)
oh_count = count(oh_effective)

found_tut_effective = tutoring_count[5] + tutoring_count[6] + tutoring_count[7]
found_oh_effective = oh_count[5] + oh_count[6] + oh_count[7]


found_tut_effective / len(tutoring_effective), found_oh_effective / len(oh_effective)
```

**Result**: Based on the proportion above we can suggest students are on average happy with the office hours than they are with the tutoring. Thus I would suggest an enhacement of the tutoring section.
